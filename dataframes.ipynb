```python
amatPlayerInfoDF.groupby("rank_season_9").size().plot(kind="pie", legend = True, autopct="%i", title="Amateur Players Rank chart", colors = ['gold', 'gainsboro', 'aquamarine'])
```

```python

HourPlot = hoursTeams.plot(by="team",kind="hist", figsize=(7,7),yticks=[1,2,3])

```

# Answer
Pie chart 1 shows that pro players are mostly ranked in the top 3 ranks of the game (Diamond, Master, Challenger)

Pie chart 2 shows that amateur players are in the lower ranks (gold, platinum, no ranking)

The amount of hours show that the amateur players have less playtime than pro players in range of 400 to 1200 hours. Pro players can have from 5000 to 10 000 hours of playtime in the game.

# Question 2: How much mouse movement is done during a game?
In this question, we shall take a look at the mouse movement data, analyze it for an amateur player and pro player.
The data will be taken from two matches each.


```python

amatMouse1 = pd.read_csv("amateurs/match_1/player_1/mouse.csv")
amatMouse1
```

```python
amatMouse2 = pd.read_csv("amateurs/match_20/player_1/mouse.csv")
amatMouse2
```

```python
amatMouseplot1 = amatMouse1["mouse_movement"].plot(figsize=(20,10), title="Mouse movement for amateur Player 2, Match 1", style="green")
amatMouseplot1.set_ylabel("Movement(DPI)")
amatMouseplot1.set_xlabel("Time(s)")
```

```python
amatMouseplot2 = amatMouse2["mouse_movement"].plot(figsize=(20,10), title="Mouse movement for amateur Player 2, Match 2", style="green")
amatMouseplot2.set_ylabel("Movement(DPI)")
amatMouseplot2.set_xlabel("Time(s)")
```

```python
proMouse1 = pd.read_csv("pros/match_4/player_0/mouse.csv")
proMouse1
```

```python
proMouse2 = pd.read_csv("pros/match_17/player_0/mouse.csv")
proMouse2
```

```python
proMouseplot1 = proMouse1["mouse_movement"].plot(figsize=(20,10), title="Mouse movement for pro Player 1, Match 1", style="red")
proMouseplot1.set_ylabel("Movement(DPI)")
proMouseplot1.set_xlabel("Time(s)")
```

```python
proMouseplot2 = proMouse2["mouse_movement"].plot(figsize=(20,10), title="Mouse movement for pro Player 1, Match 2", style="red")
proMouseplot2.set_ylabel("Movement(DPI)")
proMouseplot2.set_xlabel("Time(s)")
```

```python

figure, axis = plt.subplots(2,1)

axis[0].plot(amatMouse1["mouse_movement"] ,color="green")
axis[0].set_title("Amateur Player Mouse Movement")
axis[0].set_xlabel("Time")

axis[1].plot(proMouse1["mouse_movement"], color="red",)
axis[1].set_title("Pro Player Mouse Movement ", )
axis[1].set_xlabel("Time")



plt.show()
```

```python

figure, axis = plt.subplots(2,1)

axis[0].plot(amatMouse2["mouse_movement"] ,color="green")
axis[0].set_title("Amateur Player Mouse Movement")
axis[0].set_xlabel("Time")

axis[1].plot(proMouse2["mouse_movement"], color="red",)
axis[1].set_title("Pro Player Mouse Movement ", )
axis[1].set_xlabel("Time")

plt.show()
```

# Answer
Both players are seemingly the same in regards to their mouse movement, due to match length, there is a significant length of the match between pros and amateurs, in which both matches were by 400 to 800 seconds shorter for the pros.
